```python
# Monter Google Drive
from google.colab import drive
import cv2
import numpy as np
import os

# Monter le drive
drive.mount('/content/drive')
```

```python


# Fonction pour traiter l'image
def process_image(image_path):
    img = cv2.imread(image_path)
    height, width = img.shape[:2]

    # Si l'image n'est pas carrée, la rendre carrée avec des bordures noires
    if height != width:
        size = max(height, width)
        delta_w = size - width
        delta_h = size - height
        top, bottom = delta_h // 2, delta_h - (delta_h // 2)
        left, right = delta_w // 2, delta_w - (delta_w // 2)
        img = cv2.copyMakeBorder(img, top, bottom, left, right, cv2.BORDER_CONSTANT, value=[0, 0, 0])

    # Redimensionner à 512x512
    resized_img = cv2.resize(img, (512, 512))

    return resized_img

# Dossier des images sur Google Drive
folder_path = '/content/drive/My Drive/cnnn/cnn2'
output_folder_path = '/content/drive/My Drive/cnnn'

# S'assurer que le dossier de sortie existe
os.makedirs(output_folder_path, exist_ok=True)

# Traiter chaque image du dossier
for image_file in os.listdir(folder_path):
    image_path = os.path.join(folder_path, image_file)

    # Vérifier que le fichier est bien une image
    if image_file.lower().endswith(('.png', '.jpg', '.jpeg')):
        processed_img = process_image(image_path)

        # Enregistrer l'image traitée dans Google Drive
        output_path = os.path.join(output_folder_path, image_file)
        cv2.imwrite(output_path, processed_img)
        print(f"Image sauvegardée : {output_path}")
```
