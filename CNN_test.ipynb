```python
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
from google.colab import files
import matplotlib.pyplot as plt

uploaded_image = files.upload()

for filename in uploaded_image.keys():
    print(f'Utilisateur a téléchargé l\'image: {filename}')

model = load_model('cnn_model.h5')


img_path = list(uploaded_image.keys())[0]
img = image.load_img(img_path, target_size=(128, 128))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array /= 255.0


predictions = model.predict(img_array)
predicted_class = np.argmax(predictions[0])


class_names = ['glioma', 'glioma', 'notumor', 'pituitary']
predicted_label = class_names[predicted_class]


plt.imshow(img)
plt.axis('off')
plt.title(f'Prédiction: {predicted_label}')
plt.show()

print(f'La classe prédite est: {predicted_label}')
```
